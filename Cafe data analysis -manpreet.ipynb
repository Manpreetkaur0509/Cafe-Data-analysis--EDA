```python
df = pd.read_excel(r'Cafe_Data_MRA (14).xlsx')
def categorize_time(time_obj):
    hour = time_obj.hour
    if 6 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 18:
        return "Afternoon"
    elif 18 <= hour < 24:
        return "Evening"
    else:
        return "Night"

# Apply the categorize_time function to the 'Time' column
df['Interval'] = df['Time'].apply(categorize_time)

df.head()
# # Print the resulting DataFrame
# print(df)
```

```python
#To create a new column 'top_10_dishes' in an existing DataFrame based on the top 10 dishes by frequency in another column, you can follow these steps:

# Use the value_counts() method to count the occurrences of each dish in the existing DataFrame's 'Item Desc' column.
# Use the head(10) method to select the top 10 dishes by frequency.
# Use the isin() method to create a boolean mask that checks if each row in the 'Item Desc' column is one of the top 10 dishes.
# Assign the boolean mask to the new 'top_10_dishes' column.

top_10_dishes = df['Item Desc'].value_counts().head(10).index
df['top_10_dishes'] = df['Item Desc'].isin(top_10_dishes)
df.head()
```
